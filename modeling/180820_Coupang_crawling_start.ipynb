```python
df = pd.DataFrame(columns=["date", "name", "item", "option", "re_title", 'review', 'star'])

for i in range(1, 50):
    date = response_.xpath('/html/body/article[{}]/div[1]/div[3]/div[2]/text()'.format(i)).extract()[0]
    
    name = response_.xpath('/html/body/article[{}]/div[1]/div[2]/span/text()'.format(i)).extract()[0][:-1]

    item = response_.xpath('/html/body/article[{}]/div[1]/div[4]/text()'.format(i)).extract()[0].split(",")[0]

    option = response_.xpath('/html/body/article[{}]/div[1]/div[4]/text()'.format(i)).extract()[0].split(",")[1]
    
    if response_.xpath('/html/body/article[{}]/div[3]/@class'.format(i)).extract()[0] == 'sdp-review__article__list__headline':
        re_title = response_.xpath('/html/body/article[{}]/div[3]/text()'.format(i)).extract()
        review = response_.xpath('/html/body/article[{}]/div[4]/div/text()'.format(i)).extract()

#     if response_.xpath('/html/body/article[{}]/div[3]/@class'.format(i)).extract()[0] == 'sdp-review__article__list__review js_reviewArticleContentContainer':
#         re_title = ['none']
#         review = response_.xpath('/html/body/article[{}]/div[3]/div/text()'.format(i)).extract()

    else:
    # 왜 elif면 안되는지 모르겠음    
        re_title = ['.']
        review = response_.xpath('/html/body/article[{}]/div[3]/div/text()'.format(i)).extract()

    star = response_.xpath('/html/body/article[{}]/div[1]/div[3]/div[1]/div/@data-rating'.format(i)).extract()[0]

    df.loc[i-1] = {'date':date, 'name':name, 'item':item, 'option':option, 're_title':re_title, 'review':review, 'star':star}
    
df
```

```python
# re_title의 의미있는 문장 발라내기(re_title은 리스트 length 1이다)

display(df['re_title'][0][0])
display(df['re_title'][0][0].split('\n'))
for i in range(len(df['re_title'][0][0].split('\n'))):
    display(df['re_title'][0][0].split('\n')[i].strip())
    
# re_title 컬럼에 적용
def re_title_filter(a):
    for i in range(len(a[0].split('\n'))):
        if len(a[0].split('\n')[i].strip()) != 0:
            return a[0].split('\n')[i].strip()
```

```python
def kkma_pos_filter(a):
    """
    df['pos']를 받아 품사 필터링을 거쳐 해당 키워드 리스트 반환
    """
    from konlpy.tag import Kkma
    kkma = Kkma()
    ls = []
    for i in range(len(a)):
        #NNG(보통 명사), NNP(고유 명사), VA(형용사), XR(어근), MAG(일반 부사), UN(명사추정범주)
        if a[i][1]=='NNG' or a[i][1]=='NNP' or a[i][1]=='VA' or a[i][1]=='UN' or a[i][1]=='XR' or a[i][1]=='MAG':
            ls.append(a[i][0])
    return list(set(ls))
```
